## KMeans Algorith

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# X = numpy array containing the data
# y_true = class labels
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Call all the rows from first and second column
plt.scatter(X[:,0], X[:,1], s=12, cmap="rainbow")
```

```python
y_true
```

```python
from sklearn.cluster import KMeans

# initialize kmeans object
kmeans = KMeans(n_clusters=4, random_state=0)

# Fit the dataset in the kmeans object
kmeans.fit(X)

# Predict the clustering index for each sample
kmeans_labels = kmeans.predict(X)

# Two in one step as preceding
# kmeans.fit_predict(X)
```

```python
kmeans_labels
```

```python
plt.scatter(X[:,0], X[:,1], c=kmeans_labels, cmap="viridis")

# Get the centroid of the clusters
centers = kmeans.cluster_centers_

# Plot the centroids with black
plt.scatter(centers[:,0], centers[:,1], c="black", cmap="viridis")
```

```python
# computes minimum distances between one point and set of points
from sklearn.metrics import pairwise_distances_argmin

# Following function is custom KMeans implementation
# will get same result as using the KMeans libray
def my_Kmeans(X, n_clusters, rseed=2):
    rand_generator = np.random.RandomState(rseed)
    i = rand_generator.permutation(X.shape[0])[:n_clusters]
    centers = X[i]


    while True:
        labels = pairwise_distances_argmin(X, centers, metric="euclidean")

        new_centers = np.array([X[labels==i].mean(0) for i in range(n_clusters)])

        if np.all(centers==new_centers):
            break
        centers = new_centers

    return centers, labels
```

```python
centers, labels = my_Kmeans(X, 4)
```

```python
centers
```

```python
labels
```

```python
plt.scatter(X[:,0], X[:,1], c=labels, cmap="viridis")

# Plot the centroids with black
plt.scatter(centers[:,0], centers[:,1], c="black", cmap="viridis")
```

## K-Medoids Algorithm

```python
from sklearn_extra.cluster import KMedoids

kmedoids = KMedoids(n_clusters=4, random_state=0, metric="manhattan")

kmedoid_labels = kmedoids.fit_predict(X)
```

```python
kmedoid_labels
```

```python
plt.scatter(X[:,0], X[:,1], c=kmedoid_labels, s=12, cmap="viridis")
```

```python
from sklearn import metrics

def evaluation_metrics(X, labels_pred, labels_true, input_metric):
    silhouette_s = metrics.silhouette_score(X, labels_pred, metric=input_metric)
    random_index = metrics.adjusted_rand_score(labels_true, labels_pred)

    metrics_name = ["Silhouette Score", "Random Index"]
    values = [silhouette_s, random_index]

    result = list(zip(metrics_name, values))

    result = pd.DataFrame(result, columns = ["Metric", "Value"])

    return result
```

```python
evaluation_metrics(X, kmedoid_labels, y_true, "manhattan")
```

## Various Clustering Algorithms

```python
from sklearn import datasets

iris = datasets.load_iris()

X = iris.data

y = iris.target

target_names = iris.target_names
```

```python
target_names
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit_predict(X)
```

```python
kmedoids = KMedoids(n_clusters=3, random_state=0)
kmedoids.fit_predict(X)
```

```python
print("KMeans\n", evaluation_metrics(X, kmeans.labels_, y, "euclidean"))
print("KMedoids\n", evaluation_metrics(X, kmedoids.labels_, y, "euclidean"))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2).fit(X)

pca_2d = pca.transform(X)
```

```python
plt.title("Reference plot")
plt.scatter(pca_2d[:,0], pca_2d[:,1], c=y, cmap="viridis")
```

```python
plt.title("KMeans Result")
plt.scatter(pca_2d[:,0], pca_2d[:,1], c=kmeans.labels_, cmap="viridis")
```

```python
plt.title("KMedoids Result")
plt.scatter(pca_2d[:,0], pca_2d[:,1], c=kmedoids.labels_, cmap="viridis")
```

```python
from sklearn.cluster import AgglomerativeClustering

ward = AgglomerativeClustering(n_clusters=3, linkage="ward")
ward_pred = ward.fit_predict(X)

avg = AgglomerativeClustering(n_clusters=3, linkage="average")
avg_pred = avg.fit_predict(X)

complete = AgglomerativeClustering(n_clusters=3, linkage="complete")
complete_pred = complete.fit_predict(X)

```

```python
evaluation_metrics(X, ward_pred, y, "euclidean")
```

```python
evaluation_metrics(X, avg_pred, y, "euclidean")
```

```python
evaluation_metrics(X, complete_pred, y, "euclidean")
```

```python
from scipy.cluster.hierarchy import linkage

linkage_type = "average"

linkage_matrix = linkage(X, linkage_type)
```

```python
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(20, 12))
dendrogram(linkage_matrix)
plt.show()
```
