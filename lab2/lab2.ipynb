```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# X = numpy array containing the data
# y_true = class labels
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Call all the rows from first and second column
plt.scatter(X[:,0], X[:,1], s=12, cmap="rainbow")
```

```python
y_true
```

```python
from sklearn.cluster import KMeans

# initialize kmeans object
kmeans = KMeans(n_clusters=4, random_state=0)

# Fit the dataset in the kmeans object
kmeans.fit(X)

# Predict the clustering index for each sample
kmeans_labels = kmeans.predict(X)

# Two in one step as preceding
# kmeans.fit_predict(X)
```

```python
kmeans_labels
```

```python
plt.scatter(X[:,0], X[:,1], c=kmeans_labels, cmap="viridis")

# Get the centroid of the clusters
centers = kmeans.cluster_centers_

# Plot the centroids with black
plt.scatter(centers[:,0], centers[:,1], c="black", cmap="viridis")
```

```python
# computes minimum distances between one point and set of points
from sklearn.metrics import pairwise_distances_argmin

# Following function is custom KMeans implementation
# will get same result as using the KMeans libray
def my_Kmeans(X, n_clusters, rseed=2):
    rand_generator = np.random.RandomState(rseed)
    i = rand_generator.permutation(X.shape[0])[:n_clusters]
    centers = X[i]


    while True:
        labels = pairwise_distances_argmin(X, centers, metric="euclidean")

        new_centers = np.array([X[labels==i].mean(0) for i in range(n_clusters)])

        if np.all(centers==new_centers):
            break
        centers = new_centers

    return centers, labels
```

```python
centers, labels = my_Kmeans(X, 4)
```

```python
plt.scatter(X[:,0], X[:,1], c=labels, cmap="viridis")

# Plot the centroids with black
plt.scatter(centers[:,0], centers[:,1], c="black", cmap="viridis")
```
