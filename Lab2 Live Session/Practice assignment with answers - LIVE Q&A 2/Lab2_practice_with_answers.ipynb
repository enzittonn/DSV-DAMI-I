2. Convert age into a categorical attribute

```python
data['age_bins'] = pd.cut(x=data['age'], bins=[0, 19, 29, 39, 49, 59, 69, 79, 99], labels=['0-19', '19-29', '29-39', '39-49', '49-59', '59-69', '69-79', '79-99'])
```

```python
data
```

3.  Plot a bar chart for the attribute age_bins, in relation to the class  (y). 

```python
by_age= data.groupby(["age_bins"]).y.value_counts()
by_age.unstack().plot(kind='bar', stacked=True)
```

4. Check if there are null values

```python
data.isnull().sum()
```

5. Create a copy of the dataset called data_categorical

```python
"""
Hint: Use the copy() method. 

Note for the copy() method:

Make a copy of this object’s indices and data.

When deep=True (default), a new object will be created with a copy of the calling object’s data and indices. Modifications to the data or indices of the copy will not be reflected in the original object (see notes below).

When deep=False, a new object will be created without copying the calling object’s data or index (only references to the data and index are copied). Any changes to the data of the original will be reflected in the shallow copy (and vice versa).


"""

data_categorical = data.copy()
```

```python
data_categorical
```

6. From data_categorical drop all the numerical columns

```python
data_categorical.drop(["age", "balance", "duration", "pdays", "campaign", "previous"], axis=1, inplace=True)
```

```python
data_categorical
```

7. Use Label Encoder from sklearn to transform the categorical data into a machine learning readable from

**Label Encoding** refers to converting the labels into numeric form so as to convert it into the machine-readable form.

Reference: https://www.geeksforgeeks.org/ml-label-encoding-of-datasets-in-python/

```python
"""
Initialize the LabelEncoder Object

Use fit_transform to encode the data_categorical dataset

Hint: Use apply() in order to encode each of the categorical columns of the dataframe

"""
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data_categorical = data_categorical.apply(le.fit_transform)
data_categorical.head()
```

## KMODES

The distance metric used for K-modes is the Hamming distance. The Hamming distance (or dissimilarity) between two rows is simply the number of columns where the two rows differ.

```python
#To download it:  pip install kmodes
#link: https://pypi.org/project/kmodes/

from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
```

8. Choose the right number of clusters by running the elbow method for Kmodes

```python
cost = []
K = range(1,4)
for k in K:
    kmode = KModes(n_clusters=k, init = "Cao", n_init = 1)
    kmode.fit_predict(data_categorical)
    cost.append(kmode.cost_)


    
plt.figure(figsize=(16,8))
plt.plot(K, cost, 'bx-')
plt.xlabel('k')
plt.ylabel('Cost')
plt.title('Elbow')
plt.show()
```

9. Run k-modes with the optimal number of clusters as found by the elbow method. 

```python
km_cao = KModes(n_clusters=2, init = "Cao", n_init = 1, verbose=1)
fitClusters_cao = km_cao.fit_predict(data_categorical)
```

10. Attach the predicted labels to the original dataframe (the one called data) as a new column named Kmodes_labels

```python
KModes_labels = pd.Series(fitClusters_cao, name="Kmodes_labels")

data = pd.concat([data, KModes_labels], axis = 1)
```

```python
data
```

11. Plot a bar chart for the attribute KModes_labels, in relation to the the marital status. 

```python
by_cluster= data.groupby(["Kmodes_labels"]).marital.value_counts()
by_cluster.unstack().plot(kind='bar')
```
