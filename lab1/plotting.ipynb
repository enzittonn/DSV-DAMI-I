## Matplotlib 

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
y = [1, 2, 3, 4]
```

```python
plt.plot(y)
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
# Empty canvas
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)

# Labels
axes.set_xlabel("X axis")
axes.set_ylabel("Y axis")

# Title
axes.set_title("This is title")
```

```python
# Multiple plots in the same canvas
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax1.plot(x, y)
ax2.plot(y, x)
```

```python
# Specify how many plots in the same canvas nrows by ncols
fig, axes = plt.subplots(nrows=1, ncols=2)

plt.tight_layout()

axes[0].plot(x, y)
axes[0].set_title("plot1")
axes[1].plot(y, x)
axes[1].set_title("plot2")
```
