# Standardization

```python
# Standardization must be performed prior to PCA
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Numerical features
numerical = ["age", "trestbps", "chol", "thalach", "oldpeak"]

# Sub dataset with numerical data from the data_dropped
X = data_dropped[[c for c in data_dropped.columns if c in numerical]]
```

```python
X
```

```python
# StandardScaler object
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
# Numpy array
X
```

```python
# Transform numpy array to df
X = pd.DataFrame(X, columns=numerical)
```

```python
# Standardized data
X
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)

pca.fit(X)

x_pca = pca.transform(X)
```

```python
x_pca
```

```python
x_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=data_dropped["num"], cmap="rainbow")
plt.xlabel("1st PC")
plt.ylabel("2st PC")
```

```python
pca.components_
```

```python
comp = pd.DataFrame(pca.components_, columns=X.columns)
```

```python
comp
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(comp, cmap="plasma")
```
